Why launching near the equator:

Launching from near the equator offers some advantages for certain types of missions due to the Earth's rotational speed. The closer you are to the equator, the faster the Earth's surface moves due to its rotation. This extra velocity can be advantageous for a few reasons:

Increased Payload Capacity: Rockets launched from near the equator can take advantage of the Earth's rotational speed, allowing them to gain extra velocity when they lift off. This additional velocity can increase the payload capacity of the rocket or allow it to reach higher orbits with the same payload.

Launch Flexibility: The higher rotational speed near the equator provides more flexibility in selecting launch azimuths, allowing for a broader range of orbits to be reached effectively.

SpaceX, like many other space launch companies, often locates its rocket launch sites near coastlines for several reasons:

Safety: Launching from coastal areas allows for rockets to fly over open water, which minimizes the risk to populated areas in case of a launch failure. If something goes wrong during launch, the rocket can be intentionally terminated over the ocean, reducing the risk of damage or harm to people and property.

Accessibility: Coastal areas typically offer good transportation access via ports and roads. This makes it easier to transport rocket components and fuel to the launch site and recover the rocket's first stage, which often lands on a drone ship in the ocean after launch.

Launch azimuth: The choice of launch azimuth, or the angle at which a rocket is launched, is often determined by the desired orbital trajectory. Coastal launch sites offer flexibility in choosing the optimal launch direction to reach specific orbits.

Reduced noise pollution: Launching from coastal sites can minimize the impact of rocket noise on nearby communities compared to launching from inland sites, as the sound dissipates more effectively over the open ocean.

Environmental considerations: Coastal sites can be easier to permit and have less impact on local ecosystems compared to launching from more remote or sensitive inland areas.

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `launch_sites` dataframe called `marker_color` to store the marker colors based on the `class` value


```python
# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red

# Define a function to map class values to marker colors
def map_marker_color(class_value):
    if class_value == 1:
        return 'green'
    elif class_value == 0:
        return 'red'
    else:
        return 'blue'  # Handle other cases as needed

# Apply the function to create the marker_color column
spacex_df['marker_color'] = spacex_df['class'].apply(map_marker_color)

# Display the updated DataFrame
print(spacex_df)
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
launch_site_lat = 34.63287
launch_site_lon = -120.61075
coastline_lat = 34.63833
coastline_lon = -120.62409
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
distance_coastline
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
distance_marker = folium.Marker(
   location=[coastline_lat,coastline_lon],
    icon=DivIcon(
        icon_size=(20,20),
      icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
        )
    )
site_map.add_child(distance_marker)
site_map
```

The minimal distance required for a rocket launch site from railways, cities, and highways can vary depending on several factors, including regulatory requirements, safety considerations, and the specific needs of the launch vehicle. These distances are typically defined by space agencies, regulatory bodies, and international agreements to ensure the safety of the public and surrounding infrastructure. Here's a general overview:

Railways: The minimal distance from railways for a rocket launch site is typically defined by safety regulations and can vary by country and space agency. In the United States, for example, the Federal Aviation Administration (FAA) provides guidelines for commercial space launch safety. These guidelines require a minimal distance from railways to ensure that a rocket failure or accident does not cause harm to trains or railway operations. The specific distance may depend on factors like the type of rocket and payload.

Cities: The minimal distance from populated areas, including cities, is a critical safety concern for rocket launch sites. Safety regulations usually dictate that launch sites should be located far enough from cities to minimize the risk to human life and property in the event of a launch failure. The specific distance varies by country and is influenced by the population density and the size of the launch vehicle.

Highways: Rocket launch sites should be located at a safe distance from highways to minimize the risk to vehicular traffic in the event of a launch failure. The specific minimal distance from highways can depend on local and national regulations and the size of the rocket. Highway closures are often coordinated for launches to ensure safety.
